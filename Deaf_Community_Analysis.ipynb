Global Deaf Research Institute Data Analysis


```python
#Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
deaf =pd.read_excel('/content/Deaf Community Research Nigeria Data for Viz for Social Good V.2.xlsx')
deaf
```

```python
deaf_lang = deaf.groupby('Languages you use comfortably')['Languages you use comfortably'].count()
deaf_lang
```

```python
deaf_occupation = deaf.groupby('Current occupation status')['Current occupation status'].count()
deaf_occupation
```

```python
plt.bar(deaf_occupation.index,deaf_occupation.values)
plt.xticks(rotation=90)
plt.show()
```

```python
deaf_salary = deaf.groupby('Current occupation status')['How much money do you earn a month? (Salary) [It is in Niara currency]']
deaf_salary
```

```python
d= deaf_salary.max()
d.head()
```

```python
deaf_salary.median()
```

```python
deaf_jobtype = deaf.groupby('What kind job do you have?')['How much money do you earn a month? (Salary) [It is in Niara currency]'].count()
deaf_jobtype
```

```python
deaf_support = deaf.groupby('What did you need to support your hearing loss experience?')['What did you need to support your hearing loss experience?'].count()
deaf_support
```

```python
from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(deaf_support)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
deaf_prob = deaf.groupby('What are the three most significant challenges you believe deaf people face in your area?')['What are the three most significant challenges you believe deaf people face in your area?'].count()
```

```python
from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(deaf_prob)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Preferred communication methods for general contact and emergencies
contact_methods = deaf['Best contact method for communication'].dropna()
emergency_contact_methods = deaf['Best contact method for emergencies'].dropna()

# Plotting preferred communication methods
plt.figure(figsize=(14, 7))
contact_method_counts = contact_methods.value_counts().head(10)  # Show top 10 methods
sns.barplot(y=contact_method_counts.index, x=contact_method_counts.values, orient='h')
plt.title('Preferred Communication Methods for General Contact')
plt.xlabel('Count')
plt.ylabel('Communication Method')
plt.show()

# Plotting preferred emergency contact methods
plt.figure(figsize=(14, 7))
emergency_contact_method_counts = emergency_contact_methods.value_counts().head(10)  # Show top 10 methods
sns.barplot(y=emergency_contact_method_counts.index, x=emergency_contact_method_counts.values, orient='h')
plt.title('Preferred Communication Methods for Emergencies')
plt.xlabel('Count')
plt.ylabel('Emergency Contact Method')
plt.show()
```
